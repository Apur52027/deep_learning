## Design a neural network using nn.Module.

### The model must contain:

```python
import torch.nn as nn

# Get the number of input features from X_train_tensor
input_size = X_train_tensor.shape[1]

# Define the neural network model
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()

        self.fc2 = nn.Linear(hidden_size, output_size)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.sigmoid(out)
        return out

# Hyperparameters for the model
hidden_size = 64 # Number of neurons in the hidden layer
output_size = 1  # For binary classification (outputting a single probability)

# Instantiate the model
model = SimpleNN(input_size, hidden_size, output_size)

# Print the model architecture
print("Model Architecture:")
print(model)

# Print the total number of trainable parameters
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTotal number of trainable parameters: {total_params}")
```

## Implement a full training loop:

Forward pass

Loss computation

Backward pass

Parameter update

Gradient reset

### Requirements:

Train for at least 100 epochs.

Print loss every 10 epochs.

Store training loss history(You can pick your own Data Structure).

Explain clearly what happens in each step of the pipeline.

```python
epochs = 100
learning_rate = 0.001 # Defined in optimizer already
training_loss_history = []

for epoch in range(epochs):
    # 1. Forward pass: Compute predicted y by passing X to the model
    y_pred = model(X_train_tensor)

    # 2. Loss computation: Calculate the loss
    # Reshape y_train_tensor to match outputs shape if necessary
    loss = loss_function(y_pred,y_train_tensor.reshape(-1,1))

    # 3. Backward pass: Compute gradient of the loss with respect to model parameters
    optimizer.zero_grad()  # Gradient reset: Clear previous gradients
    loss.backward()        # Backpropagation: Compute gradients

    # 4. Parameter update: Adjust model parameters using the optimizer
    optimizer.step()

    # Store training loss history
    training_loss_history.append(loss.item())

    # Print loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

print("\nTraining complete.")
print(f"Final training loss: {training_loss_history[-1]:.4f}")
```

## Write Answer 04:

The training loop has successfully completed, and the loss has decreased over the 100 epochs, indicating that the model is learning. The final training loss achieved was **0.0907**.

Here's an explanation of what happens in each step of the training pipeline:

1.  **Forward Pass**: `outputs = model(X_train_tensor)`
    *   In this step, the input training data (`X_train_tensor`) is fed into the neural network model. The model performs a series of computations, including matrix multiplications and applying activation functions (ReLU in the hidden layer and Sigmoid in the output layer), to generate predictions (`outputs`). These `outputs` are the model's estimate of the probability that each sample belongs to the positive class.

2.  **Loss Computation**: `loss = loss_function(y_pred,y_train_tensor.reshape(-1,1))`
    *   After the forward pass, the calculated `outputs` (predictions) are compared against the true target values (`y_train_tensor`). The `loss_function` (Binary Cross-Entropy Loss) quantifies the difference between the predicted probabilities and the actual labels. A lower loss value indicates that the model's predictions are closer to the true labels.

3.  **Gradient Reset**: `optimizer.zero_grad()`
    *   Before performing backpropagation for the current iteration (epoch), it is essential to clear the gradients from the previous iteration. PyTorch, by default, accumulates gradients. If we don't reset them, the gradients would be the sum of gradients from all previous iterations, leading to incorrect and unstable parameter updates. This step ensures that we compute gradients solely based on the current batch's loss.

4.  **Backward Pass (Backpropagation)**: `loss.backward()`
    *   This is the core of training. Based on the computed loss, `loss.backward()` calculates the gradients of the loss with respect to every trainable parameter (weights and biases) in the model. This process uses the chain rule to determine how much each parameter contributed to the error, propagating the error signal backward through the network layers.

5.  **Parameter Update**: `optimizer.step()`
    *   Finally, the `optimizer` (Adam in this case) uses the gradients calculated in the backward pass to adjust the model's parameters. It moves the parameters in the direction that reduces the loss. The `learning_rate` (set to 0.001) determines the size of these steps. This iterative adjustment process is how the model learns to make more accurate predictions over time.

## Write Answer 07:

### Why must gradients be reset every epoch?
In PyTorch (and other deep learning frameworks), gradients are accumulated by default. This means that if we don't explicitly reset them, the gradients from the current batch/epoch will be added to the gradients from previous batches/epochs. This accumulation would lead to incorrect gradient values, as the optimizer would be trying to update the model parameters based on an average gradient over multiple steps, rather than the gradient calculated specifically for the current batch's loss. `optimizer.zero_grad()` clears these accumulated gradients, ensuring that each parameter update is based solely on the gradients computed from the current forward and backward pass.

### What happens if the learning rate is too high?
If the learning rate is too high, the optimizer will take very large steps during parameter updates. This can cause several issues:
1.  **Overshooting the minimum**: The optimizer might jump over the optimal minimum of the loss function, leading to oscillations around the minimum or even diverging completely.
2.  **Instability**: The loss function can become unstable, showing erratic behavior, and may not converge at all.
3.  **Divergence**: In extreme cases, a very high learning rate can cause the model's weights to grow exponentially, leading to NaN (Not a Number) values in the loss or weights, and the training process will diverge.

### What happens if the learning rate is too small?
Conversely, if the learning rate is too small, the optimizer will take very tiny steps during parameter updates:
1.  **Slow convergence**: The training process will be very slow, taking a long time to reach the minimum of the loss function. This means the model might require many more epochs to train effectively.
2.  **Getting stuck in local minima**: The model might get trapped in a suboptimal local minimum or a saddle point, as the small steps might not be sufficient to escape flatter regions of the loss landscape.
3.  **Suboptimal performance**: Even if it converges, it might settle for a suboptimal solution, not achieving the best possible performance.

### Why do we define layers inside the constructor ( `__init__` ) and not inside `forward()`?
In PyTorch's `nn.Module`:
1.  **Initialization and State Management**: Defining layers (like `nn.Linear`, `nn.ReLU`, etc.) in the `__init__` method ensures that these layers are initialized once when the model object is created. They become part of the model's state, meaning their parameters (weights and biases) are registered as `nn.Parameter` objects. This allows PyTorch to automatically track these parameters for gradient computation during the backward pass and include them in the `model.parameters()` iterator for the optimizer.
2.  **Efficiency**: If layers were defined inside the `forward()` method, a new instance of each layer would be created every time the `forward()` method is called (i.e., for every single forward pass during training or inference). This would be extremely inefficient, as it would repeatedly allocate memory and re-initialize weights and biases. It would also prevent the model from learning, as the parameters would be reset in each step, making backpropagation and parameter updates meaningless.
3.  **Modularity and Readability**: It separates the definition of the model's architecture (what layers it has) from its computation logic (how data flows through those layers), improving code organization, readability, and maintainability. The `__init__` method sets up the building blocks, and `forward()` describes how these blocks are connected.
